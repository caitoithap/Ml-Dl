Encode the label variable with numerical values in order to be able to build machine-learning models
Encode the month, dai_of_week,education,housing, loan and default attributes using the OrdinalEncoder class

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

# Sample data (you would replace this with your actual dataset)
data = {
    'month': ['jan', 'feb', 'mar', 'apr', 'may'],
    'day_of_week': ['mon', 'tue', 'wed', 'thu', 'fri'],
    'education': ['primary', 'secondary', 'tertiary', 'unknown', 'secondary'],
    'housing': ['yes', 'no', 'yes', 'yes', 'no'],
    'loan': ['no', 'yes', 'no', 'no', 'yes'],
    'default': ['no', 'yes', 'no', 'no', 'yes']
}

df = pd.DataFrame(data)

# Instantiate the OrdinalEncoder
encoder = OrdinalEncoder()

# Apply OrdinalEncoder to the columns
encoded_df = df.copy()
encoded_df[['month', 'day_of_week', 'education', 'housing', 'loan', 'default']] = encoder.fit_transform(
    df[['month', 'day_of_week', 'education', 'housing', 'loan', 'default']]
)

# Display the transformed DataFrame
print(encoded_df)
```
