```python
import numpy as np

# Chuyển quality thành nhãn binary
df["quality_binary"] = np.where(df["quality"] > 6, 1, 0)
df1 = df.copy()
df1['quality'] = df['quality_binary']
df1.drop(columns=['quality_binary'], inplace=True)
# Kiểm tra số lượng mẫu trong từng nhóm
print(df1["quality"].value_counts())
```

```python
from sklearn.model_selection import train_test_split
target = "quality"
x = df1.drop(target,axis=1)
z= df1[target]

x_train,x_test,z_train,z_test = train_test_split(x,z ,test_size= 0.2, random_state= 42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
x_train
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import classification_report
model = ExtraTreesClassifier()
model.fit(x_train,z_train)

z_predict = model.predict(x_test)

print(classification_report(z_test,z_predict))
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    "n_estimators": [100,200,300],
    "criterion" : ["gini", "entropy", "log loss"]
}

md = GridSearchCV(estimator=ExtraTreesClassifier(), param_grid=params,cv=4, scoring="accuracy",verbose=2)
md.fit(x_train,z_train)
print(md.best_params_)
print(md.best_score_)
```

```python
from lazypredict.Supervised import LazyClassifier
clf = LazyClassifier(verbose=0, ignore_warnings= True,custom_metric=None)
model, predictions = clf.fit(x_train,x_test,z_train,z_test)
print(model)
```
